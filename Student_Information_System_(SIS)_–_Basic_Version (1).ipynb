```python
#project1

students = {}
for i in range(1, 4):
    print(f"\nEnter details for Student {i}:")
    name = input("Full Name: ")
    phone = input("Phone Number: ")
    address = input("Address: ")
    courses = {}
    print("Enter details for 3 courses:")

    for c in range(1, 4):
        course_name = input(f" Course {c} Name: ")
        grade1 = float(input("  Grade 1: "))
        grade2 = float(input("  Grade 2: "))
        grade3 = float(input("  Grade 3: "))

        courses[course_name] = {
            "Grade 1": grade1,
            "Grade 2": grade2,
            "Grade 3": grade3
        }


    students[name] = {
        "Phone": phone,
        "Address": address,
        "Courses": courses
    }


print("\nn--- Student Lookup ---")
search_name = input("Enter the student's full name: ")

if search_name in students:
    print("\nStudent Details:")
    print(f"Name: {search_name}")
    print(f"Phone: {students[search_name]['Phone']}")
    print(f"Address: {students[search_name]['Address']}")

    print("\nCourses and Grades:")
    for course, grades in students[search_name]["Courses"].items():
        print(f"  {course}: {grades}")
else:
    print("Student not found.")
```
